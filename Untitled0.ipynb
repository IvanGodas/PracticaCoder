```python
def suma():
  num_uno = int(input("Ingrese un numero: "))
  num_dos = int(input("Ingrese otro numero: "))

  print(f"El resultado de la suma es {num_uno + num_dos}")


def resta():
  num_uno = int(input("Ingrese un numero: "))
  num_dos = int(input("Ingrese otro numero: "))

  print(f"El resultado de la suma es {num_uno - num_dos}")


suma()
resta()
```

```python
def suma( num_uno , num_dos):
  print(f"El resultado de la suma es {num_uno + num_dos}")


def resta( num_uno , num_dos):
  print(f"El resultado de la suma es {num_uno - num_dos}")


suma(20,30)
suma(50,60)
suma(100,300)
resta(20,56)
resta(50,34)
```

```python
def factorial(n):
  if n==0 or n==1:
    return 1
  else:
    return (n*factorial(n-1))

print(factorial(4))
```

```python
def area_rectangulo(base, altura):
  area = base * altura
  return area

area_rectangulo(15,10)
```

```python
import math
def area_circulo(radio):
  area = (radio)**2 * math.pi
  return area

area_circulo(5)

```

```python
def relacion(num1, num2):
  if num1 > num2:
    print("1")
  elif num1 < num2:
    print("-1")
  else:
    print("0")

relacion(5,10)
relacion(10,5)
relacion(5,5)
```

```python
def recortar(numero_recortar, limite_inferior, limite_superior):
  if numero_recortar < limite_inferior:
    return limite_inferior
  elif numero_recortar > limite_superior:
    return limite_superior
  else:
    return numero_recortar

recortar(15,0,10)
```

```python
def separar(numeros):
    lista = sorted(numeros)
    pares = []
    impares = []
    for num in lista:
        if num % 2 == 0:
            pares.append(num)
        else:
            impares.append(num)
        
    return pares, impares

separar([6,5,2,1,7])
```
